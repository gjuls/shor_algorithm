```python
import numpy as np
from qiskit.extensions import UnitaryGate

n = 5
N = 21
y = 4

dim = 2**n
matrix_0 = np.zeros((dim, dim))

#CREATE F(X)
for i in range(N):
    j = (i*y)%N
    matrix_0[j][i]=1
            
for ii in range(N, dim):
    matrix_0[ii][ii] = 1

temp = matrix_0       
power = 2*n
for iii in range(power):
    matrix = temp.dot(temp)
    temp = matrix

#CREATE GATE
U = UnitaryGate(temp)

U.name = "%i^%i mod %i" % ( y, 2**power, N)
c_U = U.control()
 
```
