# Periodicity

```python
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(18, 6))
params = {'text.usetex' : True, 'font.size' : 20, 'font.family' : 'serif'}
plt.rcParams.update(params)

N = 15
y = 13

xs = []
xpoints = []
ypoints = []
data = []
for i in range(20):
    data.append((y**i)%N)
    xs.append(i)
    if ((y**i)%N) == 1:
        xpoints.append(i)
        ypoints.append((y**i)%N)
plt.plot(data, color = '#4C495A')
plt.plot([xpoints[0], xpoints[1]], [ypoints[0], ypoints[1]], linestyle="dotted", color = 'grey')
#plt.scatter(xpoints, ypoints, marker = "X", facecolors="grey", alpha = 1) 
plt.xticks(xs, xs)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x; r)$")
plt.savefig('period.pdf', format = 'pdf')
```
