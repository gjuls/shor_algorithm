```python
import matplotlib.pyplot as plt

params = {'text.usetex' : True, 'font.size' : 20, 'font.family' : 'serif'}
plt.rcParams.update(params)
plt.rc('figure', figsize=(12, 6))

x = [0, 1, 2, 3, 4, 5]
XZ1_3 = [0.2657142857,0.28,0.2771428571,0.2971428571,0.2942857143,0.2971428571]
XZ3_5 = [0.2914285714,0.3057142857,0.3,0.2942857143,0.2942857143,0.2771428571]
XZ1_15 = [0.4428571429,0.4142857143,0.4228571429,0.4085714286,0.3828571429,0.4257142857]

plt.plot(x, XZ1_3, label = '1 x 3', color = '#B3BFB8')
plt.plot(x, XZ3_5, label = '3 x 5', color = '#FF7F51')
plt.plot(x, XZ1_15, label = '1 x 15', color = '#395756')
plt.legend()

plt.xlabel('Number of error')
plt.ylabel('Frequency', rotation = 90)
plt.tight_layout()

plt.savefig('ErrorsXZ.pdf', format = 'pdf')

```
